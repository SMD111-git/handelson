# assaign new cloumn from the existing one


```python
iris=pd.read_csv("Student_Marks.csv")
iris.head()
```

```python
iris.assign(speal_ratio=iris["time_study"]/iris["Marks"]).head()
```

# in the example above, we inserted a precomputed value. We can also pass in a function of one argument to be evaluated on the DataFrame being assigned to.

```python
iris.assign(sepal_ratio=lambda x: (x["time_study"] / x["Marks"])).head()
```

or, using pandas.col():

```python
iris.assign(speal_ratio=pd.col("time_study")/pd.col("Marks")).head()
```

 just experiment of graphs

```python
(
    iris.assign(
        sepal_ratio = iris["time_study"] / iris["Marks"]
    )
    .plot(
        kind="scatter",
        x="Marks",
        y="time_study"
    )
)
```
