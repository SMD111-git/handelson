# loading or creating of data


```python
import numpy as np
studentmarks=np.array([[55,60,70],
[80,60,70],
[80,70,60]],dtype=int)
print(studentmarks.dtype)
```

```python
np.save("studentmarks.npy",studentmarks)
```

```python
loading=np.load("studentmarks.npy")
print(loading)
```

```python
sumsubjects=np.sum(loading,axis=0)
print(sumsubjects)
print(sumsubjects[0] ," frist subject total")
print(sumsubjects[1] ," second subject total")
print(sumsubjects[2] ," thrid subject total")


#toating as per the students marks row vise
studentper=np.sum(loading,axis=1)
print(studentper)
print(studentper[0], "student number one")
print(studentper[1], "student number two")
print(studentper[2], "student number three")
```

#trying to get the avg and top among the student of three

```python
avgstudentscore=np.average(studentper)
print(avgstudentscore)
```

min and max amoung the studets

```python
minmarks=np.min(loading,axis=1)
print(minmarks)
maxsocre=np.max(loading,axis=1)
print(maxsocre)
```
